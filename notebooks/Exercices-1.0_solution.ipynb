Modifiez le code suivant pour afficher 'il reste VALEUR CLÉ dans le magasin', pour chaque élément CLÉ: VALEUR du dictionnaire `inventaire`.

````python
inventaire = {'papier':5, 'pain':7, 'céréale':9}

# version A: boucle for utilisant les clés du dictionnaire
for clé in inventaire:
    print('il reste', inventaire[clé], clé, 'dans le magasin.')
````

# Exercice 7
Créez une fonction qui prend en argument une liste avec tous les montants des transactions de banque que vous avez faites, et qui retourne le montant que vous avez en banque suite à ces transactions. 

On suppose que le montant disponible en banque au départ était de zéro.

Ex.: 

    total_banque([1, 25, -12, 30])
    44

    total_banque([])
    0

    total_banque(range(10))
    45

````python
# version utilsant une boucle for
def total_banque(transactions):
    solde = 0
    for montant in transactions:
        solde = solde + montant
    return solde
````

````python
# version plus simple
def total_banque(transactions):
    return sum(transactions)
````

````python
total_banque([1, 25, -12, 30])
````

````python
total_banque([])
````

````python
total_banque(range(10))
````

# Exercice 8


Pouvez-vous modifier la fonction `sport_jouer` afin qu'elle retourne `trop chaud pour jouer` si la température est plus grande que 25 ?

Vous pouvez aussi faire en sorte que la fonction `sport_jouer` retourne `aucun sport` si la température est entre 0 et 10 degrés.



````python
def sport_jouer(température):
    if température > 25:
        return 'trop chaud pour jouer'
    elif température > 10:
        return 'soccer'
    elif température < 0:
        return 'hockey'
    else:
        return 'aucun sport'
````

````python
sport_jouer(30)
````

````python
sport_jouer(7)
````

## Partie 1

Créez une fonction qui imprime une pyramide composée de symboles `*` et dont l'argument est le nombre d'étages de la pyramide

Par exemple, pour le nombre 3, nous avons:

````python
print('  *')
print(' ***')
print('*****')
````

````python
def pyramide(N):
    for i in range(N):
        print(' '*(N-i-1) + '*'*(2*i+1))
````

##Tranches

La tranche (*slice* en anglais) permet de sélectionner une partie d'une séquence. On peut l'utiliser entre autres avec des listes ou des chaînes de caractères. Alors que l'indice nous permet de sélectionner un seul élément d'une séquence, la tranche nous permet d'en sélectionner plusieurs d'un coup. Comme pour l'indice, on précise la tranche entre deux crochets ```[]```. Dans sa forme la plus commune, on précise l'indice de départ (inclusif) et l'indice de fin (exclusif) des éléments qu'on veut sélectionner, séparés par un ```:```, donc ```[d:f]```


````python
fruits = ['pomme', 'banane', 'orange', 'ananas', 'fraise', 'framboise','bleuet']
print(fruits[2])   # indice
print(fruits[4])   # indice
print(fruits[2:4]) # tranche
````

La tranche nous renvoie une nouvelle séquence composée des éléments sélectionnés. Comme le premier indice d'une séquence est le ```0```, on voit que l'élément à l'indice ```2``` est en fait le 3e élément de la liste. Et comme l'indice de fin de la tranche est exclusif, la tranche s'arrête en fait à l'élément précédent.

Il est facultatif de préciser l'indice de début et de fin. Si on omet l'indice de début, la tranche débutera au premier élément de la séquence. Si on omet l'indice de fin, elle se terminera au dernier élément. On peut même omettre les deux indices, dans lequel cas la séquence retournée sera identique à l'originale. Voyons un exemple avec une chaîne de caractères.



````python
nom = 'Rosalie'
print(nom[:4])
print(nom[4:])
print(nom[:])  # équivaut à nom[0:len(nom)]
````

Comme pour les indices, on peut utiliser des indices négatifs pour définir notre tranche. On peut aussi assigner le résultat de notre tranche à une nouvelle variable. Voici un exemple:

````python
frigo = ['orange', 'pomme', 'carotte', 'concombre']
fruits = frigo[:-2]
legumes = frigo[-2:]
print(fruits)
print(legumes)
````

On peut aussi préciser la taille des pas de notre tranche. Par défaut, la tranche parcourt la séquence de notre indice de début à notre indice de fin en faisant des pas d'un élément à la fois. Mais on peut lui demander de faire des pas plus grands. Pour ce faire, on doit préciser la taille des pas en utilisant un deuxième ```:```


````python
résultats = ['victoire', 'défaite', 'victoire', 'défaite']
victoires = résultats[0:4:2]  # ou résultats[::2]
print(victoires)
````

On peut aussi faire des pas inverses pour parcourir la séquence à l'envers en précisant un nombre de pas négatif. Voici un exemple de fonction qui utillise cette possibilité.

````python
def renverser(sequence):
    return sequence[::-1]

nom = 'Édith'
print(renverser(nom))
````

# Syndication Web

Adepte des [flux RSS](https://fr.wikipedia.org/wiki/RSS) pour suivre l'activité de vos sites web préférés?

Ça fait un peu "années 2000" mais ça s'appelle la [Syndication de contenus web](https://fr.wikipedia.org/wiki/Syndication_de_contenu).

Python peut vous aider à syndiquer vos flux (feeds).

## Objectif

On veut créer du code qui va automatiquement aller sélectionner les 5 dernières actualités affichées sur le site de Montréal-Python: http://montrealpython.org/fr/feed/

## Approche

Voici l'approche proposée

1. utiliser `feedparser`
    * `feedparser` est une _library_ Python qui n'est pas installée par défaut quand on installe Python (pas dans la _Standard Library_
    * il faut donc l'installer avec pip
        * si vous travaillez [avec Colab, installer avec la syntaxe spéciale: `!`](https://colab.research.google.com/notebooks/snippets/importing_libraries.ipynb)
            ```
            !pip install feedparser
            ```
        * si vous travaillez en local, sur votre machine, simplement
            ```
            pip install feedparser
            ```
2. importer et suivre exemple de la doc : http://packages.python.org/feedparser/
3. introspecter (et imprimer dictionnaire pour voir ses clés) au besoin
4. coder le code qui fera le traitement voulu
5. lancer le code pour confirmer son exécution correcte



## Algorithme

1. capter le flux RSS de Montréal-Python: http://montrealpython.org/fr/feed/
2. retenir le nombre d'items voulus
3. traiter les items retenus
    * créer une chaîne de caractères unique : avec les infos des derniers items modifiés (titre, URL... et éventuellement la date de dernière modification)

## Solution

Prenez le temps de coder vous-même une solution...
... ensuite vous pouvez comparer avec : [le solutionnaire](https://colab.research.google.com/drive/1vpYeC3hgnMhLN1KRaiHdULHw1qNHojTO?usp=sharing)

Pour info, notre solution ne fait que 6 lignes Python.

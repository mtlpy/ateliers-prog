## Solution verbeuse

Avec algorithmique en commentaire et un peu plus d'étapes.

```python
import feedparser
       
# capter le flux RSS
url = "http://montrealpython.org/fr/feed/"
flux = feedparser.parse(url)

# retenir le nombre d'items voulus
#items = flux['entries']     # tous
entries = flux['entries'][0:5]  # 5 derniers items car déjà triés par .updated

# traiter les items retenus
print("Les 5 dernières modifications du site de Montréal-Python:")
for e in entries:
    # print chaîne formattée (f-string)
    #print(f"* {e.link} : {e.title}, {e.published")
    print(f"* {e.link} : {e.title}")
```

## Solution simplifiée

```python
import feedparser
        
url = "http://montrealpython.org/fr/feed/"
flux = feedparser.parse(url)
        
print("Les 5 dernières modifications du site de Montréal-Python :")
for e in flux.entries[0:5]:
    print(f"* {e.link} : {e.title}")
```
